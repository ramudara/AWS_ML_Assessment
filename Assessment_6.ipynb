# Problem Statement:

6.	Using the Facial analysis feature of Amazon Rekognition, create a program that accepts an input image and outputs the number of faces found in it, as well as the characteristics of each face such as male/female, age group, smiling or not, wearing glasses, having a beard/moustache, etc. Use multiple images to cover all use cases â€“ one without any face, only 1 face, multiple faces. Submit the program, images, and their corresponding output.

#####Importing required libraries

```python
import boto3
import json
```

# *************Main Program*****************

```python
def detect_num_faces():
    client=boto3.client('rekognition', region_name='eu-west-1')
    inp_img = input("Please enter the image file name:")
    image = open(inp_img, 'rb')
    response = client.detect_faces(Image={'Bytes': image.read()}, Attributes=['ALL'])
    #response = client.index_faces(CollectionId="family_collection",Image={'Bytes': image.read()}, DetectionAttributes=['ALL'])
    num_faces = len(response['FaceDetails'])
    print("The number of faces detected in the given image equals to {}".format(num_faces))
    if num_faces == 0:
       print("No face detected in the given image")
    else:
        i = 1	
        for faceDetail in response['FaceDetails']:
            print ("Detected attributes for face {} are as follows\n".format(i))
            print('The detected face is between ' + str(faceDetail['AgeRange']['Low'])
                  + ' and ' + str(faceDetail['AgeRange']['High']) + ' years old')
            print('Here are the other attributes:')
            print(json.dumps(faceDetail, indent=4, sort_keys=True))
            i = i + 1
```

# IMAGE with ONE Face in it.

```python
detect_num_faces()
```

```python
detect_num_faces()
```
